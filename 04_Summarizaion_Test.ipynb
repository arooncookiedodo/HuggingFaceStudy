# **2. 토크나이저 및 모델 로드 : facebook/bart-large-cnn**

```python
# 토크나이저 및 모델 로드 : facebook/bart-large-cnn

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

# **3. 데이터셋 로드 : cnn_dailymail**

```python
# cnn_dailymail 데이터셋 로드

from datasets import load_dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")  # load_dataset("모델명", "버전")
```

# **4. 데이터셋 살펴보기**

```python
# dataset의 구조

dataset
```

```python
# cnn_dailymail 첫번째를 예시로 사용

example_text = dataset["train"][0]["article"]
```

```python
# 예시 데이터 원본 Print
print("Original Text:")
print(example_text)
```

```python
# 예시 데이터의 Reference Summary
reference_summary = dataset["train"][0]["highlights"]

#예시 데이터의 참조 Summary (Highlights)
print("Highlights:")
print(reference_summary)
```

# **5. 데이터 전처리 및 요약**

```python
# 예시 데이터 전처리 및 요약


#Tokenizer 및 데이터 전처리
input_ids = tokenizer.encode(example_text, return_tensors="pt", max_length=1024, truncation=True)

#요약 생성
summary_ids = model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)

#요약 Decode
generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# 생성된 요약 Print
print("Summary Text: ")
print(generated_summary)
```

# **6. ROUGE Score 평가**

```python
# 예시 데이터의 ROUGE 스코어 계산

from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = scorer.score(reference_summary, generated_summary)

```

# **7. 결과 출력**

```python
# 예시 데이터 원본 Print
print("Original Text: ")
print(example_text+ '\n')

# 생성된 요약 Print
print("generated_summary: ")
print(generated_summary + '\n')

# 참조 요약 Print
print("reference_summary : ")
print(reference_summary + '\n')

# ROUGE 점수
print("ROUGE scores: ")
print(rouge_scores)
```

추가> 훈련 데이터셋 살펴보기

```python
from datasets import load_dataset

# cnn_dailymail 데이터셋 로드
dataset = load_dataset("cnn_dailymail", "3.0.0")

# 훈련 데이터셋을 셔플하고 처음 10개의 예시를 선택
shuffled_dataset = dataset["train"].shuffle(seed=42)  # seed 값은 일관된 결과를 위해 설정
sampled_examples = shuffled_dataset.select(range(10))

# 예시 데이터 출력
for i, example in enumerate(sampled_examples):
    print(f"Example {i+1}:")
    print(example["article"])
    print("\n")
```
